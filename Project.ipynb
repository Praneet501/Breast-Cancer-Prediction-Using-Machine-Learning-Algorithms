```python
def cdf(data): #form a cumulative distribution function for data
    n = len(data)
    x = np.sort(data)
    y = np.arange(1, n+1) / n
    return x, y
```

#### E) BootStrap Replicate

```python
def bootstrap_replicate_1d(data, func):
    #Generate bootstrap replicate of 
    bs_sample = np.random.choice(data, len(data))
    return func(bs_sample)

def draw_bs_reps(data, func, size=1):
    #Draw bootstrap replicates

    bs_replicates = np.empty(size)
    
    for i in range(size):
        bs_replicates[i] = bootstrap_replicate_1d(data,func)

    return bs_replicates

bs_replicates = draw_bs_reps(cancer_data['radius_mean'], np.mean, size=10000)

conf_int = np.percentile(bs_replicates,[2.5, 97.5])
print('95% confidence interval =', conf_int, 'hours')


____ = plt.hist(bs_replicates, bins=50, density=True, alpha = .8, color = 'Orange')
____ = plt.xlabel('Radius ')
____ = plt.ylabel('P(T <= t)')
____ = plt.title('Probability Density Function')

plt.show()
```

#### B) Anova Table

```python
# Create ANOVA backbone table
data1 = [['Between Groups', '', '', '', '', '', ''], ['Within Groups', '', '', '', '', '', ''], ['Total', '', '', '', '', '', '']] 
anova_table = pd.DataFrame(data1, columns = ['Source of Variation', 'SS', 'df', 'MS', 'F', 'P-value', 'F crit']) 
anova_table.set_index('Source of Variation', inplace = True)

# calculate SSTR and update anova table
x_bar = y.mean()
SSTR = cancer_data.groupby('diagnosis').count() * (cancer_data.groupby('diagnosis').mean() - x_bar)**2
anova_table['SS']['Between Groups'] = SSTR['radius_mean'].sum()

# calculate SSE and update anova table
SSE = (cancer_data.groupby('diagnosis').count() - 1) * cancer_data.groupby('diagnosis').std()**2
anova_table['SS']['Within Groups'] = SSE['radius_mean'].sum()

# calculate SSTR and update anova table
SSTR = SSTR['radius_mean'].sum() + SSE['radius_mean'].sum()
anova_table['SS']['Total'] = SSTR

# update degree of freedom
anova_table['df']['Between Groups'] = cancer_data['diagnosis'].nunique() - 1
anova_table['df']['Within Groups'] = cancer_data.shape[0] - cancer_data['diagnosis'].nunique()
anova_table['df']['Total'] = cancer_data.shape[0] - 1
```

```python
# calculate MS
import scipy.stats as stats
anova_table['MS'] = anova_table['SS'] / anova_table['df']

# calculate F 
F = anova_table['MS']['Between Groups'] / anova_table['MS']['Within Groups']
anova_table['F']['Between Groups'] = F

# p-value
anova_table['P-value']['Between Groups'] = 1 - stats.f.cdf(F, anova_table['df']['Between Groups'], anova_table['df']['Within Groups'])

# F critical 
alpha = 0.05
# possible types "right-tailed, left-tailed, two-tailed"
tail_hypothesis_type = "two-tailed"
if tail_hypothesis_type == "two-tailed":
    alpha /= 2
anova_table['F crit']['Between Groups'] = stats.f.ppf(1-alpha, anova_table['df']['Between Groups'], anova_table['df']['Within Groups'])

# Final ANOVA Table
anova_table
```

#### (IV) t-test on samples from distribution numTests x alphaSet times for Type I errors

```python
for i in range(1,numTests+1):  
    for alpha in ahpha_set:
        # take two samples from the same population
        samp1 = np.random.choice(pop,100,replace=True)
        samp2 = np.random.choice(pop,100,replace=True)
        # test sample means
        result = stats.ttest_ind(samp1, samp2)
        # Evaluate whether Null hypothesis for TYPE I error
        if result[1] < alpha:
             sigTests.loc[counter] = [1, result[1], alpha]
        else:
            sigTests.loc[counter] = [0, result[1], alpha]
            counter += 1
```

```python
N = len(cancer_data.radius_mean)
df_a = len(cancer_data.diagnosis.unique()) - 1
df_b = len(cancer_data.area_mean.unique()) - 1
df_axb = df_a*df_b 
df_w = N - (len(cancer_data.diagnosis.unique())*len(cancer_data.area_mean.unique()))
```

```python
grand_mean = cancer_data['radius_mean'].mean()
```

```python
# Sum of Squares A
ssq_a = sum([(cancer_data[cancer_data.diagnosis == l].radius_mean.mean() - grand_mean)**2 for l in cancer_data.diagnosis])
# Sum of Squares B
ssq_b = sum([(cancer_data[cancer_data.area_mean ==l].radius_mean.mean() - grand_mean)**2 for l in cancer_data.area_mean])
# Sum of Squares T
ssq_t = sum((cancer_data.radius_mean - grand_mean)**2)
```

```python
M = cancer_data[cancer_data.diagnosis == 0]
B = cancer_data[cancer_data.diagnosis == 1]
M_area_means = [M[M.area_mean == d].radius_mean.mean() for d in M.area_mean]
B_area_means = [B[B.area_mean == d].radius_mean.mean() for d in B.area_mean]
ssq_w = sum((B.radius_mean - B_area_means)**2) +sum((M.radius_mean - M_area_means)**2)
# Interaction
ssq_axb = ssq_t - ssq_a - ssq_b - ssq_w
```

```python
ms_a = ssq_a/df_a
ms_b = ssq_b/df_b
ms_axb = ssq_axb/df_axb
ms_w = ssq_w/df_w
```

```python
f_a = ms_a/ms_w
f_b = ms_b/ms_w
f_axb = ms_axb/ms_w
```

```python
p_a = stats.f.sf(f_a, df_a, df_w)
p_b = stats.f.sf(f_b, df_b, df_w)
p_axb = stats.f.sf(f_axb, df_axb, df_w)
```

```python
results = {'sum_sq':[ssq_a, ssq_b, ssq_axb, ssq_w],
           'df':[df_a, df_b, df_axb, df_w],
           'F':[f_a, f_b, f_axb, 'NaN'],
            'PR(>F)':[p_a, p_b, p_axb, 'NaN']}

columns=['sum_sq', 'df', 'F', 'PR(>F)']

aov_table1 = pd.DataFrame(results, columns=columns,
                          index=['diagnosis', 'area_mean', 
                          'diagnosis:area_mean', 'Residual'])
```

```python
def eta_squared(aov):
    aov['eta_sq'] = 'NaN'
    aov['eta_sq'] = aov[:-1]['sum_sq']/sum(aov['sum_sq'])
    return aov

def omega_squared(aov):
    mse = aov['sum_sq'][-1]/aov['df'][-1]
    aov['omega_sq'] = 'NaN'
    aov['omega_sq'] = (aov[:-1]['sum_sq']-(aov[:-1]['df']*mse))/(sum(aov['sum_sq'])+mse)
    return aov


eta_squared(aov_table1)
omega_squared(aov_table1)
print(aov_table1)
```

```python
def model_building(model, X_train, X_test, y_train, y_test):
    """
    
    Model Fitting, Prediction And Other stuff
    return ('score', 'accuracy_score', 'predictions' )
    """
    
    model.fit(X_train, y_train)
    score = model.score(X_train, y_train)
    predictions = model.predict(X_test)
    accuracy = accuracy_score(predictions, y_test)
    
    return (score, accuracy, predictions)  
```
